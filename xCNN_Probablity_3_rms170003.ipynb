```
import matplotlib.pyplot as plt
import random
import numpy as np
```

```
KNOWLEDGE_ANSWERS = [x for x in range(100)]

TEST_BATCHSIZE = 100
```

```
def mean_std(TEST_BATCHSIZE):
  MEAN_GUESSED_COUNT =[]
  STD_GUESSED_COUNT =[]
  for k in KNOWLEDGE_ANSWERS:
    guess = np.zeros(TEST_BATCHSIZE)
    for tb in range(TEST_BATCHSIZE):
      for _ in range(100-k):
        guess[tb] += (int(random.random()>0.75))
    MEAN_GUESSED_COUNT.append(np.mean(guess))
    STD_GUESSED_COUNT.append(np.std(guess))
  fig, ax = plt.subplots()
  ax.plot(KNOWLEDGE_ANSWERS, MEAN_GUESSED_COUNT, label="MEANS")
  ax.plot(KNOWLEDGE_ANSWERS, STD_GUESSED_COUNT, label="STANDARD DEVIATION")
  ax.set_xlabel('Score by knowledge')
  ax.set_title('Batch Size '+str(TEST_BATCHSIZE))
  ax.legend()

  plt.show()
```

```
for TEST_BATCHSIZE in [5,10,100,500]:
  mean_std(TEST_BATCHSIZE)
```
